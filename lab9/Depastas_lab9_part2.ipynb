### Εκπαίδευση του Μοντέλου

Χρησιμοποιώντας το πακέτο Scikit-Learn θα εκπαιδεύσουμε τον αλγόριθμο SVM πάνω στα δεδομένα του προηγούμενου προβλήματος, χρησιμοποιώντας γραμμικό πυρήνα.

```python
model = SVC(kernel='linear', C=1E10)
model.fit(X, y)
```

Με την παρακάτω μέθοδο θα εμφανιστεί στο σχήμα που ακολουθεί η βέλτιστη γραμμή διαχωρισμού των δύο κατηγοριών, καθώς και τα περιθώριά της. Τα training examples που ακουμπούν πάνω στα όρια ονομάζονται Support Vectors.

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plot_svc_decision_function(model);
```

Στο ακόλουθο σχήμα, η διαδικασία επαναλαμβάνεται για δύο datasets που περιλαμβάνουν 60 και 120 training examples αντίστοιχα. 
<ul>
<li><b>Ερώτηση 5:</b> Τι παρατηρείτε σχετικά με τη βέλτιστη γραμμή που διαχωρίζει τις δύο κατηγορίες και τα περιθώριά της;</li>
<li><b>Ερώτηση 6:</b>Τι αρκεί να κάνει ο αλγόριθμος για να ταξινομήσει νέα δεδομένα που θα λάβει στην είσοδό του;</li>
</ul>

```python
def plot_svm(N=10, ax=None):
    X, y = make_blobs(n_samples=200, centers=2,
                      random_state=0, cluster_std=0.60)
    X = X[:N]
    y = y[:N]
    model = SVC(kernel='linear', C=1E10)
    model.fit(X, y)
    
    ax = ax or plt.gca()
    ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
    ax.set_xlim(-1, 4)
    ax.set_ylim(-1, 6)
    plot_svc_decision_function(model, ax)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)
for axi, N in zip(ax, [60, 120]):
    plot_svm(N, axi)
    axi.set_title('N = {0}'.format(N))
```

<b>Ερώτηση 5:</b> Τι παρατηρείτε σχετικά με τη βέλτιστη γραμμή που διαχωρίζει τις δύο κατηγορίες και τα περιθώριά της;

Αρχικά, βλέπουμε πως και στις δύο περιπτώσεις η διαχωριστική γραμμή και οι δύο περιθωριακές γραμμές της ικανοποιούν τις απαιτήσεις που τέθηκαν σε προηγούμενη ερώτηση. Βλέπουμε ότι μάλιστα, για να βελτιστοποιηθεί το περιθώριο, οι περιθωριακές γραμμές εφάπτονται σε δειγματικά στοιχεία από τις 2 κλάσεις. Κάποια από αυτά τα σημεία είναι support vectors και αυτά (και μόνον αυτά) ορίζουν τη βέλτιστη διαχωριστική ευθεία γραμμή αλλά και τις περιθωριακές.


<b>Ερώτηση 6:</b>Τι αρκεί να κάνει ο αλγόριθμος για να ταξινομήσει νέα δεδομένα που θα λάβει στην είσοδό του;

Ο αλγόριθμος αρκεί να εξετάσει το πρόσημο της συνάρτησης $g(\vec{x})$ για ένα διάνυσμα εισόδου $\vec{x}$, καθώς η εξίσωση της ευθείας διαχωρισμού είναι: $g(\vec{x})=0$. Επομένως υπολογίζεται το $g(\vec{x})$ και αν  $g(\vec{x}) > 0$ η είσοδος $\vec{x}$ ταξινομείται στην κλάση/κατηγορία 1 (κόκκινα) αφού βρίσκεται πάνω από τη διαχωριστική ευθεία γραμμή, ενώ αν $g(\vec{x}) < 0$, τότε η ταξινόμηση γίνεται στην κλάση 2 (κίτρινα). Όταν $g(\vec{x}) = 0$ υπάρχουν δύο επιλογές, είτε το δειγματικό στοιχείο ταξινομείται τυχαία σε μια εκ των δύο κλάσεων είτε δε γίνεται καμία ταξινόμηση του δειγματικού στοιχείου.

### Μη Γραμμικά Διαχωρίσιμες Κατηγορίες: Kernel SVM

Στην περίπτωση που οι κατηγορίες του προβλήματος δεν είναι γραμμικά διαχωρίσιμες, ο αλγόριθμος SVM χρησιμοποιεί μη γραμμικούς πυρήνες (kernels) για να προβάλει τα δεδομένα εκπαίδευσης σε μεγαλύτερη διάσταση όπου μπορεί να τα καταστήσει γραμμικά διαχωρίσιμα. Μία τέτοια περίπτωση δεδομένων εκπαίδευσης παρουσιάζεται στο επόμενο σχήμα.

```python
X, y = make_circles(100, factor=.1, noise=.1)

clf = SVC(kernel='linear').fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plot_svc_decision_function(clf, plot_support=False);
```

Είναι φανερό ότι τα δεδομένα εκπαίδευσης δεν μπορούν να διαχωριστούν γραμμικά. Στο επόμενο σχήμα φαίνεται η προβολή τους στις τρεις διαστάσεις με τη χρήση ενός μη γραμμικού πυρήνα.

<li><b>Ερώτηση 8:</b> Δώστε ένα παράδειγμα γραμμικού ορίου που θα μπορούσε να διαχωρίσει τις δύο κατηγορίες.</li>

```python
r = np.exp(-(X ** 2).sum(1))
```

```python
def plot_3D(elev=30, azim=30, X=X, y=y):
    ax = plt.subplot(projection='3d')
    ax.scatter3D(X[:, 0], X[:, 1], r, c=y, s=50, cmap='autumn')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('r')

interact(plot_3D, elev=[-90, 90], azip=(-180, 180),
         X=fixed(X), y=fixed(y));
```

<b>Ερώτηση 7:</b> Τι είδους πυρήνας χρησιμοποιήθηκε;

Όπως φαίνεται από το σχήμα του πυρήνα, αυτός είναι μη γραμμικός και μάλιστα όπως είδαμε στο part 1, είναι πυρήνας ακτινικής συνάρτησης βάσης (RBF - radial basis function). Από το cell κώδικα `r = np.exp(-(X ** 2).sum(1))` προκύπτει ότι χρησιμοποιήθηκε ως κέντρο το $\vec{w} = (0,0)$ και ως ακτίνα $\sigma = \frac{\sqrt{2}}{2}$, όπου η ακτινική συνάρτηση βάσης δίνεται από την:
$h(\vec{x})=\exp \left(-\frac{\left\|\vec{x}-\vec{w}\right\|^{2}}{2 \sigma^{2}}\right)=\exp \left(-\frac{\sum_{i=1}^{d}\left(x_{i}-w_{i}\right)^{2}}{2 \sigma^{2}}\right) = exp(-||\vec{x}||^2)$

```python
0.521 to max gia class 0
0.998 to max gia class 1 kai min 0.884
```

```python
import pandas as pd
a = pd.DataFrame(data={'r': r, 'label': y})
class0_min = a[a['label']==0]['r'].min()
class0_max = a[a['label']==0]['r'].max()
class1_min = a[a['label']==1]['r'].min()
class1_max = a[a['label']==1]['r'].max()

print(f'The plane equation can be r=value where value is between [{class0_max},{class1_min}]')
```

Η εύρεση του πυρήνα που μπορεί να διαχωρίσει με τον καλύτερο τρόπο τα δεδομένα απαιτεί, συνήθως, πολλούς υπολογισμούς.

<b>Ερώτηση 9:</b> Ποια μέθοδος μπορεί να βοηθήσει σε αυτό και πώς; (hint: <a href="https://en.wikipedia.org/wiki/Kernel_trick">Kernel Trick</a>)

Η μέθοδος που μπορεί να βοηθήσει στην εύρεση πυρήνα που διαχωρίζει με τον καλύτερο τρόπο τις δυο κλάσεις λέγεται kernel trick και ουσιαστικά υπαγορεύει πως για να βρεθεί η διαχωριστική επιφάνεια των δύο κλάσεων δεν απαιτείται να προσδιοριστεί η συνάρτηση ακτινικής βάσης $h(\vec{x})$. Αυτό γιατί μπορεί να βρεθεί σχέση της $g(x)$, όπου η διαχωριστική επιφάνεια δίνεται από την εξίσωση $g(x)=0$ η οποία δεν περιέχει την $h(\vec{x})$ μόνη της αλλά ως εσωτερικό γινόμενο $h(\vec{x})^Τ \cdot h(\vec{x}') = Κ(\vec{x},\vec{x}')$. Στην προηγούμενη σχέση $Κ(\vec{x},\vec{x}')$ είναι μια συνάρτηση πυρήνα που μετρά την ομοιότητα μεταξύ δύο δειγματικών στοιχείων εισόδου $\vec{x},\vec{x}' \in X$ και $K: X \times X \rightarrow \mathbb{R}$.

Συνεπώς, η βελτιστοποίηση που θέλουμε να γίνει για την εύρεση του καλύτερου διαχωριστικού επιπέδου μπορεί να γίνει με τον προσδιορισμό της συνάρτησης πυρήνα αντί τον προσδιορισμό της ακτινικής συνάρτησης βάσης, μειώνοντας σημαντικά την πολυπλοκότητα του προβλήματος από τετραγωνική σε γραμμική.
