<h1><b><i>Ο αλγόριθμος Προσομοιωμένης Ανόπτησης</i> (<i>Simulated Annealing</i>)</b></h1>

Στην άσκηση αυτή θα μελετήσετε τον αλγόριθμο της <i>Προσομοιωμένης Ανόπτησης</i> (<i>Simulated Annealing</i>), περισσότερες πληροφορίες για τον οποίο μπορείτε να βρείτε <a href="https://en.wikipedia.org/wiki/Simulated_annealing">εδώ</a>. Θα χρησιμοποιήσετε το <b>έτοιμο</b> πρόγραμμα που δίνεται στη συνέχεια. Το πρόγραμμα αυτό επιλύει το πρόβλημα του <i>Περιπλανώμενου Πωλητή</i> (<i>Travelling Salesman Problem</i>, <i>TSP</i>), περισσότερες πληροφορίες για το οποίο μπορείτε να βρείτε <a href="https://en.wikipedia.org/wiki/Travelling_salesman_problem">εδώ</a>. Αρχικά, το πρόβλημα λύνεται χρησιμοποιώντας τον αλγόριθμο <i>Simulated Annealing</i> και, στη συνέχεια, χρησιμοποιώντας τον αλγόριθμο <a href="https://en.wikipedia.org/wiki/Hill_climbing"><i>Hill Climbing</i></a>. Η επίλυση του προβλήματος γίνεται για <b>150 πόλεις</b>.

```python
#!/usr/bin/env python

""" Traveling salesman problem solved using Simulated Annealing.
http://www.physics.rutgers.edu/~haule/681/src_MC/python_codes/salesman.py
"""
from scipy import *
from pylab import *
from math import floor

def Distance(R1, R2):
    return sqrt((R1[0] - R2[0]) ** 2 + (R1[1] - R2[1]) ** 2)

def TotalDistance(city, R):
    dist = 0
    for i in range(len(city) - 1):
        dist += Distance(R[city[i]], R[city[i + 1]])
    dist += Distance(R[city[-1]], R[city[0]])
    return dist
    
def reverse(city, n):
    nct = len(city)
    nn = floor((1 + ((n[1] - n[0]) % nct)) / 2) # half the length of the segment to be reversed
    # the segment is reversed in the following way n[0]<->n[1], n[0]+1<->n[1]-1, n[0]+2<->n[1]-2,...
    # Start at the ends of the segment and swap pairs of cities, moving towards the center.
    for j in range(nn):
        k = (n[0] + j) % nct
        l = (n[1] - j) % nct
        city[k], city[l] = city[l], city[k]
    
def transpt(city, n):
    nct = len(city)
    newcity=[]
    # Segment in the range n[0]...n[1]
    for j in range((n[1] - n[0]) % nct + 1):
        newcity.append(city[(j + n[0]) % nct])
    # is followed by segment n[5]...n[2]
    for j in range((n[2] - n[5]) % nct + 1):
        newcity.append(city[(j + n[5]) % nct])
    # is followed by segment n[3]...n[4]
    for j in range( (n[4] - n[3]) % nct + 1):
        newcity.append(city[(j + n[3]) % nct])
    return newcity

def Plot(city, R, dist):
    # Plot
    Pt = [R[city[i]] for i in range(len(city))]
    Pt += [R[city[0]]]
    Pt = array(Pt)
    title('Total distance='+str(dist))
    plot(Pt[:,0], Pt[:,1], '-o')
    show()

    ###########################################################################################
    ################################# Simulated Annealing Case ################################
    ###########################################################################################

if __name__=='__main__':

    ncity = 150        # Number of cities to visit
    maxTsteps = 100    # Temperature is lowered not more than maxTsteps
    Tstart = 0.2       # Starting temperature - has to be high enough
    fCool = 0.9        # Factor to multiply temperature at each cooling step
    maxSteps = 100 * ncity     # Number of steps at constant temperature
    maxAccepted = 10 * ncity   # Number of accepted steps at constant temperature

    Preverse = 0.5      # How often to choose reverse/transpose trial move

    # Choosing city coordinates
    R=[]  # coordinates of cities are choosen randomly
    for i in range(ncity):
        R.append([rand(), rand()])
    R = array(R)

    # The index table -- the order the cities are visited.
    city = range(ncity)
    city = list(city)
    # Distance of the travel at the beginning
    dist = TotalDistance(city, R)

    # Stores points of a move
    n = zeros(6, dtype = int)
    nct = len(R) # number of cities
    
    T = Tstart # temperature

    Plot(city, R, dist)
    
    for t in range(maxTsteps):  # Over temperature

        accepted = 0
        for i in range(maxSteps): # At each temperature, many Monte Carlo steps
            
            while True: # Will find two random cities sufficiently close by
                # Two cities n[0] and n[1] are choosen at random
                n[0] = int((nct)*rand())     # select one city
                n[1] = int((nct-1)*rand())   # select another city, but not the same
                if (n[1] >= n[0]): n[1] += 1   #
                if (n[1] < n[0]): (n[0],n[1]) = (n[1],n[0]) # swap, because it must be: n[0]<n[1]
                nn = (n[0]+nct -n[1]-1) % nct  # number of cities not on the segment n[0]..n[1]
                if nn>=3: break
        
            # We want to have one index before and one after the two cities
            # The order hence is [n2,n0,n1,n3]
            n[2] = (n[0] - 1) % nct  # index before n0  -- see figure in the lecture notes
            n[3] = (n[1] + 1) % nct  # index after n2   -- see figure in the lecture notes
            
            if Preverse > rand(): 
                # Here we reverse a segment
                # What would be the cost to reverse the path between city[n[0]]-city[n[1]]?
                de = Distance(R[city[n[2]]],R[city[n[1]]]) + Distance(R[city[n[3]]],R[city[n[0]]]) - Distance(R[city[n[2]]],R[city[n[0]]]) - Distance(R[city[n[3]]],R[city[n[1]]])
                
                if de<0 or exp(-de/T)>rand(): # Metropolis
                    accepted += 1
                    dist += de
                    reverse(city, n)
            else:
                # Here we transpose a segment
                nc = (n[1]+1+ int(rand()*(nn-1)))%nct  # Another point outside n[0],n[1] segment. See picture in lecture nodes!
                n[4] = nc
                n[5] = (nc+1) % nct
        
                # Cost to transpose a segment
                de = -Distance(R[city[n[1]]],R[city[n[3]]]) - Distance(R[city[n[0]]],R[city[n[2]]]) - Distance(R[city[n[4]]],R[city[n[5]]])
                de += Distance(R[city[n[0]]],R[city[n[4]]]) + Distance(R[city[n[1]]],R[city[n[5]]]) + Distance(R[city[n[2]]],R[city[n[3]]])
                
                if de<0 or exp(-de/T)>rand(): # Metropolis
                    accepted += 1
                    dist += de
                    city = transpt(city, n)
                    
            if accepted > maxAccepted: break

        # Plot
        Plot(city, R, dist)
            
        print("Step=%d, T=%10.5f , distance= %10.5f , accepted steps= %d" %(t, T, dist, accepted))
        T *= fCool             # The system is cooled down
        if accepted == 0: break  # If the path does not want to change any more, we can stop
        
    ###########################################################################################
    ############################# Repeating for Hill Climbing Case ############################
    ###########################################################################################
    print("Repeating for Hill Climbing Case")
    
    # The index table -- the order the cities are visited.
    city = range(ncity)
    city = list(city)
    # Distance of the travel at the beginning
    dist = TotalDistance(city, R)

    # Stores points of a move
    n = zeros(6, dtype=int)
    nct = len(R) # number of cities
    
    T = Tstart # temperature

    Plot(city, R, dist)
    
    for t in range(maxTsteps):  # Over temperature

        accepted = 0
        for i in range(maxSteps): # At each temperature, many Monte Carlo steps
            
            while True: # Will find two random cities sufficiently close by
                # Two cities n[0] and n[1] are choosen at random
                n[0] = int((nct)*rand())     # select one city
                n[1] = int((nct-1)*rand())   # select another city, but not the same
                if (n[1] >= n[0]): n[1] += 1   #
                if (n[1] < n[0]): (n[0],n[1]) = (n[1],n[0]) # swap, because it must be: n[0]<n[1]
                nn = (n[0]+nct -n[1]-1) % nct  # number of cities not on the segment n[0]..n[1]
                if nn>=3: break
        
            # We want to have one index before and one after the two cities
            # The order hence is [n2,n0,n1,n3]
            n[2] = (n[0]-1) % nct  # index before n0  -- see figure in the lecture notes
            n[3] = (n[1]+1) % nct  # index after n2   -- see figure in the lecture notes
            
            if Preverse > rand(): 
                # Here we reverse a segment
                # What would be the cost to reverse the path between city[n[0]]-city[n[1]]?
                de = Distance(R[city[n[2]]],R[city[n[1]]]) + Distance(R[city[n[3]]],R[city[n[0]]]) - Distance(R[city[n[2]]],R[city[n[0]]]) - Distance(R[city[n[3]]],R[city[n[1]]])
                
                if de<0: # Hill Climbing
                    accepted += 1
                    dist += de
                    reverse(city, n)
            else:
                # Here we transpose a segment
                nc = (n[1]+1+ int(rand()*(nn-1)))%nct  # Another point outside n[0],n[1] segment. See picture in lecture nodes!
                n[4] = nc
                n[5] = (nc+1) % nct
        
                # Cost to transpose a segment
                de = -Distance(R[city[n[1]]],R[city[n[3]]]) - Distance(R[city[n[0]]],R[city[n[2]]]) - Distance(R[city[n[4]]],R[city[n[5]]])
                de += Distance(R[city[n[0]]],R[city[n[4]]]) + Distance(R[city[n[1]]],R[city[n[5]]]) + Distance(R[city[n[2]]],R[city[n[3]]])
                
                if de<0: # Hill Climbing
                    accepted += 1
                    dist += de
                    city = transpt(city, n)
                    
            if accepted > maxAccepted: break

        # Plot
        Plot(city, R, dist)
            
        print("Step=%d, T=%10.5f , distance= %10.5f , accepted steps= %d" %(t, T, dist, accepted))
        T *= fCool             # The system is cooled down
        if accepted == 0: break  # If the path does not want to change any more, we can stop
```

## Q1
<i>Να διατυπώσετε το πρόβλημα TSP</i>

<p></p>

Το πρόβλημα του Travelling Salesman (Travelling Salesman Problem) μπορεί να διατυπωθεί περιγραφικά ως εξής:

Ένας πωλητής πρέπει να εκκινήσει από μια συγκεκριμένη πόλη και να επισκεφτεί κάθε άλλη πόλη σε μια επικράτεια ακριβώς μία φορά και να επιστρέψει τελικά στην αρχική πόλη. Αυτό πρέπει να το επιτύχει με το ελάχιστο δυνατό κόστος, ήτοι υπάρχει μια συνάρτηση κόστους (πχ. απόσταση, κόστος μετακίνησης, χρόνος μετακίνησης κτλ). Ως δεδομένο λαμβάνεται το κόστος μεταξύ δύο κόμβων (ζευγών πόλεων δηλαδή) όπως η απόστασή τους λόγου χάρη και αναζητείται η διαδρομή με το μικρότερο δυνατό κόστος.

Για παράδειγμα, έστω ότι ο πωλητής έχει ως βάση του την πόλη Α και θέλει να επισκεφθεί 3 πόλεις (Β, Γ και Δ) από μία ακριβώς φορά πριν επιστρέψει στην πόλη Α. Τότε έχοντας ως δεδομένο τις αποστάσεις σε ζεύγη όλων των πόλεων μπορούν να βρεθούν τα πιθανά μονοπάτια. Σε αυτή την περίπτωση, λόγω του μικρού αριθμού πόλεων είναι δυνατό να βρεθούν όλα τα πιθανά μονοπάτια και να υπολογιστούν τα συνολικά distances για αυτά ούτως ώστε να βρεθεί το βέλτιστο. Πχ για το μονοπάτι $path_1 = Α \rightarrow B \rightarrow Γ \rightarrow Δ \rightarrow Α$, το κόστος του μονοπατιού/διαδρομής είναι $Cost(path = path_1)  = dist(A,B) + dist(B,Γ) + dist(Γ,Δ) + dist(Δ,Α)$. Για το συγκεκριμένο παράδειγμα, όλα τα δυνατά paths με αρχική πόλη την Α είναι 6 σε αριθμό και είναι τα παρακάτω:
- $path_1 = Α \rightarrow B \rightarrow Γ \rightarrow Δ \rightarrow Α$
- $path_2 = Α \rightarrow B \rightarrow Δ \rightarrow Γ \rightarrow Α$
- $path_3 = Α \rightarrow Γ \rightarrow Β \rightarrow Δ \rightarrow Α$
- $path_4 = Α \rightarrow Γ \rightarrow Δ \rightarrow Β \rightarrow Α$
- $path_5 = Α \rightarrow Δ \rightarrow Β \rightarrow Γ \rightarrow Α$
- $path_6 = Α \rightarrow Δ \rightarrow Γ \rightarrow Β \rightarrow Α$

Γενικώς, για το αντίστοιχο πρόβλημα με $n$ πόλεις συνολικά υπάρχουν $(n-1)!$ διαφορετικά paths από μια οποιαδήποτε αρχική πόλη στον εαυτό της περνώντας ακριβώς μία φορά από όλες τις υπόλοιπες $(n-1)$ πόλεις. Αυτό πχ στην περίπτωση με τις 4 πόλεις που είδαμε σημαίνει $(4-1)! = 3! = 6$ paths όπως δείξαμε.

Ωστόσο, όταν ένα TSP πρόβλημα είναι συμμετρικό, τότε το κόστος μετάβασης από έναν κόμβο/πόλη Χ σε έναν άλλο Υ είναι το ίδιο με το κόστος μετάβασης από τον κόμβο Υ στον κόμβο Χ. Για παράδειγμα, όταν το κόστος είναι η χιλιομετρική απόσταση τότε προφανώς το κόστος (απόσταση) είναι το ίδιο για τις μεταβάσεις $ X \rightarrow Y$ και $Y \rightarrow X$. Επομένως, για το προηγούμενο παράδειγμα με τις 4 πόλεις, έχουμε τις εξής ισοδυναμίες από θέμα κόστους για τα paths

- $Cost(path_1) = Cost(path_6) = dist(A,B) + dist(B,Γ) + dist(Γ,Δ) + dist(Δ,Α)$. Είναι επίσης εμφανές αυτό αν διαβάσουμε το μονοπάτι ανάποδα, δηλαδή από το τέλος προς την αρχή πως μιλάμε ουσιαστικά για την ίδια διαδρομή ανάποδα όμως.
- $Cost(path_2) = Cost(path_4)$
- $Cost(path_3) = Cost(path_5)$

Επομένως τελικά έχουμε ένα search space μεγέθους $\frac{(n-1)!}{2} = \frac{n!}{2n}$, ενώ τα συνολικά δυνατά paths που εκκινούν από οποιαδήποτε δυνατή πόλη ($n$ επιλογές) είναι $n \cdot (n-1)! = n!$. Ωστόσο το συνολικό κόστος ενός path είναι ανεξάρτητο του κόμβου/πόλης εκκίνησης.

Είναι εμφανές πως το search space παίρνει πολύ μεγάλες τιμές πολύ γρήγορα, ακόμα και για σχετικά μικρές τιμές του αριθμού κόμβων/πόλεων $n$.
Πχ.
- $n=5$: search space size $=12$
- $n=10$: search space size $=181,440$
- $n=20$: search space size $> 6 \cdot 10^{16}$

Επομένως, είναι μη εφικτό υπολογιστικά να βρούμε όλα τα πιθανά permutations για έναν αριθμό πόλεων της τάξης του 20 ή περισσότερο και να υπολογίσουμε το κόστος όλων των πιθανών paths ώστε να βρεθεί το βέλτιστο. Αυτός είναι ο λόγος που χρησιμοποιούνται κατάλληλοι αλγόριθμοι για την εύρεση μονοπατιών με μικρό κόστος.

## Q3

Ο αλγόριθμος Simulated Annealing είναι μια επέκταση της ιδέας του stochastic Hill-Climbing αλγορίθμου, ο οποίος με τη σειρά του είναι μια τροποποίηση του iterative Hill-Climbing αλγορίθμου.

Ο αλγόριθμος αυτός διαλέγει μια αρχική λύση και επαναλαμβάνει μέχρι να ικανοποιηθεί ένα κριτήριο τερματισμού (εξωτερικό loop), ενώ στο εσωτερικό loop επαναλαμβάνει για έναν αριθμό επαναλήψεων:
βρίσκει έναν γείτονα της τρέχουσας λύσης και αξιολογεί τον γείτονα με βάση τη συνάρτηση κόστους. Εφόσον προκύπτει καλύτερη λύση τότε ανανεώνει τη λύση με τη νέα, αλλιώς το κάνει με μια πιθανότητα. Αυτό επιτρέπει στον αλγόριθμο να ξεφεύγει από τοπικά ελάχιστα. Τέλος, επαναυπολογίζονται παραάμετροι όπως η ενέργεια αλλά και η παράμετρος που αφορά στον αριθμό των γειτόνων προς εξέταση σε κάθε επίπεδο ενέργειας.
Η ενέργεια μετά από κάθε ανανέωση μειώνεται με αποτέλεσμα να οδηγούμαστε σε λύσεις πιο κοντά στον τρόπο που λειτουργεί ο Hill-Climbing. Επομένως, ο Simulated Annealing έχει κατά βάση έναν τρόπο να ξεφεύγει από τοπικά ακρότατα στην αρχή και να καταλήγει παρόμοια με τον Hill-Climbing, δηλαδή να γίνονται αποδεκτές από ένα σημείο και έπειτα μόνο καλύτερες λύσεις. Αυτός είναι και ο λόγος που μπορεί να δώσει καλύτερες λύσεις από τον Hill-Climbing.
