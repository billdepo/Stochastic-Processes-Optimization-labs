<p align="justify">Αρχικά, θα εγκαταστήσετε και θα φορτώσετε τις απαραίτητες βιβλιοθήκες.</p>


```
!pip install numpy
!pip install matplotlib
!pip install pandas
!pip install sklearn

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

Στη συνέχεια, θα φορτώσετε τα δεδομένα εκπαίδευσης που παρέχονται από το αρχείο
 <b><i>data2.csv</b></i>.

```
datas = pd.read_csv('data2.csv')
print(datas)
```

Τώρα, θα δημιουργήσετε τις μεταβλητές εισόδου και εξόδου για τους αλγορίθμους μηχανικής μάθησης από τα δεδομένα εκπαίδευσης που παρέχονται.

```
X = datas.iloc[:, 1:2].values
y = datas.iloc[:, 2].values

print("Input")

print(X)

print("Labels")

print(y)
```

Στη συνέχεια, θα εκπαιδεύσετε το μοντέλο <b><i>linear regression</i></b>, χρησιμοποιώντας τα παραπάνω δεδομένα εκπαίδευσης.

```
lin = LinearRegression()
lin.fit(X, y)
```

<p align="justify">Να εκτελέσετε τα παρακάτω τμήματα κώδικα για τις τιμές της παραμέτρου <b><i>degree</b></i> {<i>2, 3, 4, 5, 8, 10, 12, 15</i>} και να καταγράψετε τα διαγράμματα που παρέχονται ως έξοδοι. <br>

**<h4> <b><i>degree</b></i> = 2 </h4>**

```
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
poly.fit(X_poly, y)
lin2 = LinearRegression()
lin2.fit(X_poly, y)

plt.scatter(X, y, color = 'blue', label = 'Training Examples')

plt.plot(X, lin.predict(X), color = 'red', label = 'Linear Regression Line')
plt.plot(X, lin2.predict(poly.fit_transform(X)), color = 'green', label = 'Polynomial Regression Line')
plt.legend()
plt.show()
```

Με την πολυωνυμική παλινδρόμηση (polynomial regression), βλέπουμε ότι καθώς αυξάνεται ο βαθμός (degree) του πολυωνύμου, η προσαρμοσμένη καμπύλη στα δεδομένα εκπαίδευσης τείνει να πηγαίνει πιο κοντά στα δειγματικά στοιχεία εκπαίδευσης (παρατηρήσεις) μειώνοντας ολοένα και περισσότερο το MSE. Αν αυξηθεί αρκετά ο βαθμός τότε η καμπύλη θα περνάει (τέμνει) από όλα τα σημεία, έχοντας MSE κοντά στο 0, ωστόσο αυτό το μοντέλο δε θα γενικεύει καλά σε νέες παρατηρήσεις.

Όσον αφορά το χρόνο εκπαίδευσης καθώς ο βαθμός αυξάνεται, παρατηρούμε ότι και αυτός αυξάνεται αφού αυξάνοντας το βαθμό του μοντέλου κατά 1, αυξάνονται οι παράμετροι προς εκτίμηση κατά 1.

<h3><b><i>Ερώτημα 2</b></i></h3>

Σας παρέχεται, επίσης, το αρχείο <b><i>data2b.csv</i></b>, το οποίο είναι το αρχείο <b><i>data2.csv</i></b> με μία παραπάνω εγγραφή, την <b><i>(9, 38, 0.3)</i></b>. Να εκτελέσετε το παρακάτω πρόγραμμα για την τιμή της παραμέτρου <b><i>degree = 15</i></b>. Πόσο μεγάλη είναι η αλλαγή που παρατηρείτε στη μορφή της γραμμής παλινδρόμησης σε σχέση με την αντίστοιχη περίπτωση του ερωτήματος 1; Τι συμπέρασμα μπορείτε να βγάλετε για τον αλγόριθμο <b><i>polynomial regression</i></b>;

```
datas = pd.read_csv('data2b.csv')
print(datas)

X = datas.iloc[:, 1:2].values
y = datas.iloc[:, 2].values

lin = LinearRegression()
lin.fit(X, y)

poly = PolynomialFeatures(degree=15)
X_poly = poly.fit_transform(X)
poly.fit(X_poly, y)
lin2 = LinearRegression()
lin2.fit(X_poly, y)

plt.scatter(X, y, color = 'blue', label = 'Training Examples')

plt.plot(X, lin.predict(X), color = 'red', label = 'Linear Regression Line')
plt.plot(X, lin2.predict(poly.fit_transform(X)), color = 'green', label = 'Polynomial Regression Line')
plt.legend()
plt.show()
```

Παρατηρούμε ότι η μορφή της καμπύλης έχει αλλάξει αισθητά λόγω της νέας παρατήρησης που προστέθηκε. Αυτό γιατί ουσιαστικά αυτή η παρατήρηση "τραβάει" την καμπύλη προς εκείνην μιας και το μοντέλο προσπαθεί για μεγάλο βαθμό (degree) να πάει όσο κοντύτερα στις παρατηρήσεις μπορεί. Η μεγάλη τιμή της εξαρτημένης μεταβλητής για αυτήν την παρατήρηση αλλοιώνει, λοιπόν, σημαντικά τη μορφή της καμπύλης.
Συμπεραίνουμε, επομένως, πως στην περίπτωση του polynomial regression, απαιτείται προσοχή στην επιλογή του βαθμού για αποφυγή overfitting στα δεδομένα εκπαίδευσης.
