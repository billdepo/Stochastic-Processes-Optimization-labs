- {6, 7, 8}: Κλειστή

Για τις καταστάσεις έχουμε:
- Επαναληπτικές (recurrent states): {1, 2, 6, 7, 8}
- Παροδικές/μεταβατικές (transient states): {3, 4, 5}

Επαναληπτικές λέγονται οι καταστάσεις αν η διαδικασία επιστρέφει στις καταστάσεις αυτές άπειρες φορές στο διηνεκές (απείρως επισκέψιμες καταστάσεις).
Παροδικές/μεταβατικές λέγονται οι καταστάσεις για τις οποίες μετά από πεπερασμένα βήματα μεταβάσεων η διαδικασία δεν επιστρέφει σε αυτές.

Ανοικτές λέγονται οι κλάσεις (υποσύνολα του χώρου καταστάσεων του μοντέλου Markov) που επιτρέπουν έξοδο προς άλλη κλάση.
Κλειστές λέγονται οι κλάσεις που δεν επιτρέπουν έξοδο προς άλλη κλάση, και οι στατιστικές ιδιότητες της αλυσίδας μετά την παρέλευση μεταβατικού χρόνου περιορίζονται σε αυτές τις κλάσεις.

## Αλυσίδα ξεκινά από κατάσταση 1

Η κατάσταση 1 αντιστοιχεί στην κατάσταση Α του σχήματος που παρουσιάστηκε από τον προσομοιωτή

Μιας και ξεκινάμε από την κλειστή κλάση {1, 2} τότε η διαδικασία παραμένει στην κλειστή αυτή κλάση. Άρα θα βρούμε την αναλλοίωτη κατανομή της κλειστής κλάσης {1, 2}.
Επικεντρώνουμε την προσοχή μας, λοιπόν, στον πίνακα πιθανοτήτων μετάβασης που αντιστοιχεί στις καταστάσεις της κλάσης αυτής, δηλαδή τις καταστάσεις 1 και 2:

\begin{equation}

P' =

\begin{bmatrix}
    1/2 & 1/2\\
    1/4 & 3/4\\
\end{bmatrix}
\end{equation}

\begin{equation}
    \begin{cases}
      π = π \cdot P'\\
      π_1 + π_2 = 1
    \end{cases}

    \Rightarrow

    \begin{cases}
      [π_1 \ π_2] = [π_1 \ π_2] \cdot \begin{bmatrix}
                                        1/2 & 1/2\\
                                        1/4 & 3/4\\ \end{bmatrix}\\
      π_1 + π_2 = 1
    \end{cases}

    \Rightarrow

    \begin{cases}
      [π_1 \ π_2] = [π_1/2 + π_2/4  \ \  π_1/2 + 3π_2/4]\\
      π_1 + π_2 = 1
    \end{cases}

    \Rightarrow

    \begin{cases}
       \begin{cases}
            π_1 = π_1/2 + π_2/4\\
            π_2 = π_1/2 + 3π_2/4
        \end{cases}\\
      π_1 + π_2 = 1
    \end{cases}

\Rightarrow
\end{equation}

\begin{equation}
\Rightarrow
    \begin{cases}
       \begin{cases}
           2 \cdot π_1 = π_2\\
            π_2 = 2 \cdot π_1
        \end{cases}\\
      π_1 + π_2 = 1
    \end{cases}

    \Rightarrow

    \begin{cases}
      π_2 = 2 \cdot π_1\\
      π_1 + π_2 = 1
    \end{cases}

    \Rightarrow

    \begin{cases}
      π_1 = 1/3\\
      π_2 = 2/3
    \end{cases}

\end{equation}

Τελικά, η αναλλοίωτη κατανομή της αλυσίδας είναι:

\begin{equation}
    π = [π_1 \ π_2 \ π_3 \ π_4 \ π_5 \ π_6 \ π_7 \ π_8] = [\frac{1}{3} \ \frac{2}{3} \ 0 \ 0 \ 0 \ 0 \ 0 \ 0]
\end{equation}

## Αλυσίδα ξεκινά από κατάσταση 8

Η κατάσταση 8 αντιστοιχεί στην κατάσταση H του σχήματος που παρουσιάστηκε από τον προσομοιωτή

Μιας και ξεκινάμε από την κλειστή κλάση {6, 7, 8} τότε η διαδικασία παραμένει στην κλειστή αυτή κλάση. Άρα θα βρούμε την αναλλοίωτη κατανομή της κλειστής κλάσης {6, 7, 8}.
Επικεντρώνουμε την προσοχή μας, λοιπόν, στον πίνακα πιθανοτήτων μετάβασης που αντιστοιχεί στις καταστάσεις της κλάσης αυτής, δηλαδή τις καταστάσεις 6, 7 και 8:

\begin{equation}

P' =

\begin{bmatrix}
    0 & 1/2 & 1/2\\
    1/2 & 0 & 1/2 \\
    1/2 & 1/2 & 0 \\
\end{bmatrix}
\end{equation}

Παρατηρούμε ότι όλες οι στήλες του πίνακα πιθανοτήτων μετάβασης αθροίζουν στη μονάδα, άρα ο πίνακας είναι διπλά στοχαστικός, το οποίο σημαίνει πως έχουμε ίση πιθανότητα να βρεθούμε στην κάθε κατάσταση εκ των {6, 7, 8} της αλυσίδας (όταν έχει παρέλθει το μεταβατικό φαινόμενο). Άρα για την αναλλοίωτη κατανομή της αλυσίδας ισχύει:

\begin{equation}
    π = [π_1 \ π_2 \ π_3 \ π_4 \ π_5 \ π_6 \ π_7 \ π_8]  = [0 \ 0 \ 0 \ 0 \ 0 \ \frac{1}{3} \ \frac{1}{3} \ \frac{1}{3}]
\end{equation}

## Να θεωρήσετε την περίπτωση στην οποία η αλυσίδα ξεκινάει από την κατάσταση 1. Να υποθέσετε ότι σε κάποιο βήμα, η αλυσίδα βρίσκεται στην κατάσταση 2 και ότι έχει παρέλθει το μεταβατικό φαινόμενο. Ποιος είναι ο αναμενόμενος αριθμός βημάτων που απαιτούνται για την πρώτη επιστροφή της αλυσίδας στην κατάσταση 2; Να απαντήσετε στο ερώτημα χωρίς τη βοήθεια υπολογιστή.

Η κατάσταση 2 είναι μία recurrent state (επαναληπτική κατάσταση), δηλαδή η διαδικασία επιστρέφει στην κατάσταση αυτή άπειρες φορές εις το διηνεκές.

Επομένως, ο χρόνος που απαιτείται για την πρώτη επιστροφή στην κατάσταση 2 της αλυσίδας, είναι ίσος με το μέσο χρόνο επιστροφής (mean recurrence time) $E[T_2(1)]$, όπου το όρισμα $k=1$ δηλώνει ότι έχουν προηγηθεί $k-1=0$ κύκλοι από την κατάσταση i=2 (δείκτης $T_i$) στον εαυτό της. Ο μέσος χρόνος επιστροφής είναι ο μέσος αριθμός βημάτων για την επιστροφή (κύκλο) από την κατάσταση i στην i αν έχουν προηγηθεί k-1 κύκλοι.

Ισχύει: $E[T_2(1)] = \frac{1}{π_2}$ Άρα με βάση την τιμή $π_2$ που βρέθηκε νωρίτερα:

$E[T_2(1)] = \frac{1}{π_2} = \frac{1}{\frac{2}{3}} = \frac{3}{2} = 1.5$

Τελικά, αναμένουμε λοιπόν την πρώτη επιστροφή στην κατάσταση 2 (πρώτος κύκλος) μετά από 1.5 βήματα.
