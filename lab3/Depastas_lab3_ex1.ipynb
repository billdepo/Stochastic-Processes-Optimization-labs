<h1><b>Άσκηση 1</b></h1>
<p align="justify">Στο ακόλουθο σχήμα φαίνεται η κάτοψη ενός σπιτιού με πέντε δωμάτια: <i>Κουζίνα (<b>Κ</b>)</i>, <i>Βιβλιοθήκη (<b>Β</b>)</i>, <i>Σαλόνι (<b>Σ</b>)</i>, <i>Υπνοδωμάτιο (<b>Υ</b>)</i> και <i>Μπάνιο (<b>Μ</b>)<i>, καθώς και οι πόρτες που τα συνδέουν. Ένα έντομο που ζει στο σπίτι διασχίζει κάθε βράδυ τυχαία μια από τις πόρτες του δωματίου στο οποίο βρίσκεται και παραμένει στο δωμάτιο που οδηγεί η πόρτα μέχρι το επόμενο βράδυ. Αρχικά το έντομο βρίσκεται στο <i>Μπάνιο</i>. Αν {X<sub>n</sub>}<sub>n</sub> είναι η μαρκοβιανή αλυσίδα που περιγράφει τη θέση του εντόμου μετά από n βράδια:<p>
<ul>
<li>Να περιγράψετε το χώρο καταστάσεων X της μαρκοβιανής αλυσίδας.</li>
<li>Να προσδιορίσετε τον πίνακα πιθανοτήτων μετάβασης P της μαρκοβιανής αλυσίδας.</li>
</ul>

<img src="https://raw.githubusercontent.com/nkostopoulos/StochasticsLabPublic/master/lab3/ex1.png"></img>

Ο χώρος καταστάσεων της Μαρκοβιανής αλυσίδας είναι το σύνολο όλων των πιθανών καταστάσεων (δωματίων) στα οποία μπορεί να βρεθεί το έντομο. Άρα, ο χώρος καταστάσεων είναι ο:
$X = \{Β, Κ, Σ, Υ, Μ\}$

Το έντομο κινούμενο μέσα στο σπίτι, επιλέγει κάθε φορά, όντας σε μία κατάσταση, την επόμενη κατάσταση με τυχαίο τρόπο. Αυτό υποδεικνύει πως όντας σε μια δεδομένη κατάσταση, το έντομο μπορεί να μεταβεί σε οποιαδήποτε γειτονική (άμεσα προσβάσιμη δηλαδή) κατάσταση ή να παραμείνει στην τρέχουσα κατάσταση με ίση πιθανότητα και όλες αυτές οι πιθανότητες θα αθροίζουν στη μονάδα (1).

Θεωρούμε εδώ το εξής mapping για τις καταστάσεις:
- 0 -> Β
- 1 -> Κ
- 2 -> Σ
- 3 -> Υ
- 4 -> Μ

Οπότε ο πίνακας μετάβασης θα είναι ο:

\begin{equation}

P =

\begin{bmatrix}
    1/3 & 1/3 & 1/3 & 0 & 0\\
    1/3 & 1/3 & 1/3 & 0 & 0\\
    1/5 & 1/5 & 1/5 & 1/5 & 1/5\\
    0 & 0 & 1/2 & 1/2 & 0\\
    0 & 0 & 1/2 & 0 & 1/2\\
\end{bmatrix}
\end{equation}
